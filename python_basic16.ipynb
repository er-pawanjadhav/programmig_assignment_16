```python
# question 2
def radians_to_degrees(n):
    p = 180/3.14
    return n*p
a = 20

print(round(radians_to_degrees(a),1))
```

```python
# question 3
def is_curzon(n):
    f = 2**n
    f = f+1
    l = 2*n
    l = l+1
    if f%l==0:
        return True
    else:
        return False

n = int(input())
print(is_curzon(n))
    
```

```python
# question 4
def area_of_hexagon(s):
    r = 3**0.5
    r = 3*r
    s = s*s
    a = s*r
    a = a/2
    return round(a,1)
n = int(input())
print(area_of_hexagon(n))
```

```python
# question 5

def binary(n):
    l = []
    i = 0;
    while (n > 0):
        l.append(n % 2) 
        n = int(n / 2)
        i += 1
    for j in range(i - 1, -1, -1):
        print(l[j], end = "")
        
n = 5
binary(n)
 
```
